```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Linear
def linear(x):
  return x
```

```python
# Sigmoid
def sigmoid(x):
  a = []
  for i in x:
    a.append(1/(1+math.exp(-i)))
  return a
```

```python
# Softmax
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
# Hyperbolic tangent
def tanh(x, derivative=False):
  if derivative == True:
    return (1-(x**2))
  return np.tanh(x)
```

```python
# ReLU
def relu(x):
  b = []
  for i in x:
    if i <0:
      b.append(0)
    else:
      b.append(i)
  return b
```

```python
# Leaky ReLU
def lr(x):
  b = []
  for i in x:
    if i<0 :
      b.append(i/10)
    else:
      b.append(i)
  return b
```

```python
# Creating an interval
x = np.arange(-3.,3.,0.1)
```

```python
lin = linear(x)
sig = sigmoid(x)
soft = softmax(x)
tanh = tanh(x)
relu = relu(x)
leaky_relu = lr(x)
# Swish
swish = sig*x
```

```python
line_0, = plt.plot(x,lin, label="Linear")
line_1, = plt.plot(x,sig, label="Sigmoid")
line_2, = plt.plot(x,soft, label="Softmax")
line_3, = plt.plot(x,tanh, label="Tanh")
line_4, = plt.plot(x,relu, label="ReLU")
line_5, = plt.plot(x,leaky_relu, label="Leaky ReLU")
line_6, = plt.plot(x,swish, label="Swish")

plt.legend(handles=[line_0,line_1,line_2,line_3,line_4,line_5,line_6],bbox_to_anchor=(1.1, 1.05))
plt.axhline(y=0,color="k")
plt.axvline(x=0, color="k")
plt.show()
```
